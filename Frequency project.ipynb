```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import scipy.io

path = "~/location/"
num = "6"
```

```python
voice = path + "v" + num
digital = path + "d" + num

y, sr = librosa.load(voice)
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=8000)
plt.figure(figsize=(8, 3.2))
librosa.display.specshow(librosa.power_to_db(S,ref=np.max),y_axis='mel', fmax=8000,x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram - voice')
plt.tight_layout()
plt.show()
y_d, sr_d = librosa.load(digital)
T = librosa.feature.melspectrogram(y=y_d, sr=sr_d, n_mels=128,fmax=8000)
plt.figure(figsize=(8, 3.2))
librosa.display.specshow(librosa.power_to_db(T,ref=np.max),y_axis='mel', fmax=8000,x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram - digital')
plt.tight_layout()
plt.show()
```

```python
rate, data = scipy.io.wavfile.read(voice)
print(rate)
print(len(data))
f, t, Sxx = scipy.signal.spectrogram(data,rate)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
rate_d, data_d = scipy.io.wavfile.read(digital)
f_d, t_d, Sxx_d = scipy.signal.spectrogram(data_d,rate_d)
plt.pcolormesh(t_d, f_d, Sxx_d)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
